## 引入

```python
import matplotlib.pyplot as plt
```

### - 最简单的画图

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])
plt.show()
```

### - 改变线条的颜色

```python
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], color='r') # color或者c, 'r'代表红色
plt.show()
```

颜色表 

| character	|  color |
|---|---| 
|'b'| blue |
|'g'| green |
|'r'| red |
|'c'| cyan |
|'m'| magenta |
|'y'| yellow |
|'k'| black |
|'w'| white |

### - 画多个图

```python
plt.subplot(2, 1, 1) # 行，列，index
plt.plot([1, 2, 3, 4], [1, 2, 3, 4])

plt.subplot(2, 1, 2)
plt.plot([1, 2, 3, 4], [1, 2, 3, 4], color='blue')

plt.show()
```
