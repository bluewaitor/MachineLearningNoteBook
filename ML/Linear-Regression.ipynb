# 线性回归

线性回归是机器学习中最简单的一种算法。它用来预测自变量x的因变量y。这样说起来可能有点绕。简单点来说，比如有个方程`y=ax+b`，其中，`x`是自变量，`y`是因变量。线性回归的最终目的是求出`a`和`b`。

学数学的时候，一般都是已知`a`和`b`的值，根据`x`的值求`y`的值。比如`y=2x+3`，已知`x=3`，求`y`。这个问题很简单，我们一眼就可以看出`y=9`。

但是在机器学习中，我们会有一大堆`(x, y)`的数据，我们需要根据这一大堆数据，找到一个方程，使得当有一个新数据`x`进来的时候，我们可以预测出最符合的`y`。举ng视频里的例子。房价预测问题，我们只考虑一个变量：`房子大小`，这里房子大小就相当于`x`, `y`相当于房价。当我们想买知道一套房子的价格的时候，只需要知道房子大小，就可以推算出房价。当然，房价的影响因素还有很多，我们现在先不考虑。


首先，我们把数据引入进来，并画图看一下

```python
import numpy as np
import matplotlib.pyplot as plt
X, Y = np.loadtxt('../Data/ex1data1.txt', delimiter=',', unpack=True)

plt.xlabel('x')
plt.ylabel('y')
plt.scatter(X, Y)
plt.plot([5, 25], [5, 25])
plt.show()
```
